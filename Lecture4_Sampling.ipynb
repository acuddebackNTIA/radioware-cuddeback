```python
import numpy as np
import matplotlib.pyplot as plt
```

# Interpolation Example
Consider the continuous time signal $f(n) = cos(\frac{2\pi n}{T})$ with a period of T = 16

```python
T_cos = 16
def f(_n):
    return np.cos(2*np.pi*(1/T_cos)*_n)
```

The code below plots the sampling of this continuous signal for a given sample rate. Run the cell below several times, incrementing the sample rate by 1 each time. What is the relationship between the sample rate and the quality of the interpolation?

```python
# discrete time sampling of continuous signal
n_cycles = 3; n_samples_per_cycle = 1 # CHANGE THIS VALUE AND RUN THE CELL - must be an integer
n = np.linspace(0, n_cycles*T_cos, (n_cycles*n_samples_per_cycle)+1); x = f(n)

# pseudo-continuous signal for comparison 
t_cont = np.linspace(0, n_cycles*T_cos, (n_cycles*100)+1)

# plotting the sample
plt.figure(figsize=(18,6)); plt.subplot(211)
s = plt.stem(n, x, label="Discrete Time Sampling", linefmt='r', markerfmt='C3o')
plt.plot(t_cont, f(t_cont), label="Continuous Time Signal")
plt.ylabel("Signal $x[n]$"); plt.xlabel("Sample $n$")
plt.title('Cosine Sampled with %i Samples Per Cycle' %n_samples_per_cycle); plt.legend()

# plotting the interpolation
# NOTE: I HAVEN'T ACTUALLY APPLIED THE FILTER TO INTERPOLATE IT  
# but this is the basic idea - students can see the interpolation get better as the sample rate increases
plt.subplot(212); plt.xlabel("Sample $n$"); plt.ylabel("Signal $x[n]$")
plt.title('Interpolated Signal Sampled with %i Samples Per Cycle' %n_samples_per_cycle)
plt.plot(t_cont, f(t_cont), label="Continuous Time Signal"); plt.plot(n, x,  'r.-', markersize=12, label='"Interpolation"')
plt.legend(); plt.subplots_adjust(top = 0.99, bottom=0.01, hspace=.5, wspace=0.4); plt.show()
```
